```python
import pandas as pd
import matplotlib as mpl #additional plotting funcitonality
mpl.rcParams['figure.dpi'] = 400 #high resolution figures
df_clean_2 = pd.read_csv(r'C:\Users\gabri\OneDrive\Documentos\Data Science Projects with Python\df_clean_2_01.csv')
```

```python
df_clean_2['EDUCATION_CAT'] =  'none'
```

```python
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
cat_mapping = {1: 'graduate school', 2: 'university', 3: 'high school', 4: 'others'}
```

```python
df_clean_2['EDUCATION_CAT'] = df_clean_2['EDUCATION'].map(cat_mapping)
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
edu_ohe = pd.get_dummies(df_clean_2['EDUCATION_CAT'])
edu_ohe.head(10)
```

```python
df_with_ohe = pd.concat([df_clean_2, edu_ohe], axis=1)
df_with_ohe[['EDUCATION_CAT', 'graduate school', 'high school', 'university', 'others']].head(10)
```

```python
df_with_ohe.to_csv(r'C:\Users\gabri\OneDrive\Documentos\Data Science Projects with Python\Chapter_1_cleaned_data.csv', index=False)
```
