## Exercise 3.01

```python
X = df[features_response].iloc[:,:-1].values
y = df[features_response].iloc[:,-1].values
print(X.shape,y.shape)
```

```python
from sklearn.feature_selection import f_classif
[f_stat, f_p_value] = f_classif(X, y)
```

```python
f_test_df = pd.DataFrame({'Feature':features_response[:-1], 'F_statistic':f_stat, 'p_value':f_p_value})
f_test_df.sort_values('p_value')
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
selector = SelectPercentile(f_classif, percentile=20)
```

```python
selector.fit(X, y)
```

```python
best_feature_ix = selector.get_support()
best_feature_ix
```

```python
features = features_response[:-1]
```

```python
best_features = [features[counter] for counter in range(len(features)) if best_feature_ix[counter]]
best_features
```

## Exercise 3.02

```python
overall_default_rate = df['default payment next month'].mean()
overall_default_rate
```

```python
df['PAY_1'].value_counts()
```

```python
group_by_pay_mean_y = df.groupby('PAY_1').agg({'default payment next month':np.mean})
group_by_pay_mean_y
```

```python
axes = plt.axes()
axes.axhline(overall_default_rate, color = 'red')
group_by_pay_mean_y.plot(marker='x', legend=False, ax=axes)
axes.set_ylabel('Proportion of credit defaults')
axes.legend(['Entire dataset', 'Groups of PAY_1'])
```

```python
pos_mask = y == 1
neg_mask = y == 0
```

```python
axes = plt.axes()
axes.hist(df.loc[neg_mask, 'LIMIT_BAL'], edgecolor='blue', color='white')
axes.hist(df.loc[pos_mask, 'LIMIT_BAL'], alpha=0.75, edgecolor=None, color='red')
axes.tick_params(axis='x', labelrotation=45)
axes.set_xlabel('Credit limit (NT$)')
axes.set_ylabel('Number of accounts')
axes.legend(['Not defaulted', 'Defaulted'])
axes.set_title('Credit limits by response variable')
```

```python
df['LIMIT_BAL'].max()
```

```python
bin_edges = list(range(0,850000,50000))
print(bin_edges)
```

```python
mpl.rcParams['figure.dpi'] = 400
axes = plt.axes()
axes.hist(df.loc[neg_mask, 'LIMIT_BAL'], bins=bin_edges, density=True, edgecolor='black', color='white')
axes.hist(df.loc[pos_mask, 'LIMIT_BAL'], bins=bin_edges, density=True, alpha=0.5, edgecolor=None, color='black')
axes.tick_params(axis='x', labelrotation=45)
axes.set_xlabel('Credit limit (NT$)')
axes.set_ylabel('Proportion of accounts')
y_ticks = axes.get_yticks()
axes.set_yticklabels(np.round(y_ticks*50000,2))
axes.legend(['Not defaulted', 'Defaulted'])
axes.set_title('Normalized distributions of credit limits by response variable')
```
